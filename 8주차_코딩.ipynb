17. 어린이/성인/노인 구분

```python
def AgeDiff(Age):
  if Age < 16:
    return '어린이'
  elif Age < 50:
    return '성인'
  else:
    return '노인'

df_k['AgeDiff'] = df_k['Age'].apply(AgeDiff)
df_k
```

18. 좌석등급과 나이구분에 따른 생존율 구해보기

```python
sns.barplot(data=df_k, x=df_k['Pclass'], y=df_k['Survived'], hue=df_k['AgeDiff'])
plt.show()
```
